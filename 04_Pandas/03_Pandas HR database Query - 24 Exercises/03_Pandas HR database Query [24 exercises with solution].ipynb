## 22. Write a Pandas program to calculate minimum, maximum and mean salary from employees file.

```python
# Solution 1:

print('Minimum Salary: {}'.format(employees_data['salary'].min()))
print('Maximum Salary: {}'.format(employees_data['salary'].max()))
print('Mean Salary: {}'.format(employees_data['salary'].mean()))
```

```python
# Solution 2: New Learning

print(employees_data.agg({'salary': ['min', 'max', 'mean']}))
```
